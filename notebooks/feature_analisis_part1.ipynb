```python
import urllib.parse
import urllib.request
```

```python
# https://www.uniprot.org/help/api_idmapping
def query_uniprot(my_query) :
    url = 'https://www.uniprot.org/uploadlists/'

    params = {
    'from': 'GENENAME',#'ACC+ID',
    'to': 'ID',# #'ACC+ID',
    'format': 'tab',
    'query': str(my_query)

#     'query': 'PF3D7_0523200 PF3D7_1040700 PF3D7_1478700 PF3D7_0201000 PF3D7_1346200'#str(my_query)
    }

    data = urllib.parse.urlencode(params)
    data = data.encode('utf-8')
    req = urllib.request.Request(url, data)
    with urllib.request.urlopen(req) as f:
        response = f.read()
    my_response = response.decode('utf-8')
#     print(response.decode('utf-8'))
    return my_response

```
